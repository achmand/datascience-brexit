## Data Visualisation

### Major Brexit Events Extracted from News Agencies 

```r
# load cleaned/augmented events dataset as df
df.events <- data.frame(read.csv(paste("data/cleansed/", kEventsFile)))

# install/load libraries  
lubripack("knitr", "IRdisplay", "kableExtra") # to show dataframe as table

# display brexit key events
kable(df.events) %>%
  kable_styling(full_width = F) %>%
  column_spec(1, bold = T, border_right = T) %>%
  column_spec(2, width = "5em") %>%
  column_spec(3, width = "80em") %>%
  as.character() %>%
  display_html()
```
