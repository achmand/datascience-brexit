### Descriptive Statistics (Properties of the datasets)

```r
##########################################################
# Descriptive Statistics Functions
##########################################################

# install/load libraries  
lubripack("e1071") 

# outputs desc stats for the specified parameters
stats.desc <- function (data, asset) {
    
    # print asset
    print(paste("Descriptive Stats for", asset))
    
    # Measures of Central Tendency 
    summary <- summary(data) # get basic stats
    
    # Dispersion Stats
    summary["Var."] <- var(data) # variance 
    summary["Std."] <- sd(data) # standard deviation
    summary["Skewness"] <- skewness(data) # skewness
    summary["Kurtosis"] <- kurtosis(data) # kurtosis

    # return descriptive stats 
    return (summary)
}
```

### Statistical Inference (Hypothesis Testing)

The two hypotheses which will be tested across various financial markets, in particular foreign exchange market, crypto currency market, equity market and commodities market are as follows:  

```r
##########################################################
# Preparing events for our model (event windowing)
##########################################################

# install/load libraries  
lubripack("broom", "dplyr") 

# there are various studies using different estimation periods such as: 60, 100, 250, 300, etc
# we choose 60 since there are many overlapping events and any events which overlap will be removed
estimation.period <- 60

# get events which have at least 60 days difference between them 
df.eventsmodel <- df.events[df.events$days_diff >= estimation.period | is.na(df.events$days_diff), ]  

# label each event for better identification 
names(df.eventsmodel)[1] = "event"
df.eventsmodel$event <- paste("event_", seq.int(nrow(df.eventsmodel)), sep="")

# we will investigate short horizons assuming that the market is efficient 
# there are various different event windows in this case study we will choose 2 days 
# set event windows (b:before event, a:after event)
df.eventsmodel$tb3 <- df.eventsmodel$event_date - 3 # event at t-3 (pre-event)
df.eventsmodel$tb2 <- df.eventsmodel$event_date - 2 # event at t-2 (pre-event)
df.eventsmodel$tb1 <- df.eventsmodel$event_date - 1 # event at t-1 (pre-event)
df.eventsmodel$ta1 <- df.eventsmodel$event_date + 1 # event at t+1 (post-event)
df.eventsmodel$ta2 <- df.eventsmodel$event_date + 2 # event at t+2 (post-event)
df.eventsmodel$ta3 <- df.eventsmodel$event_date + 3 # event at t+3 (post-event)

# estimation period start & end date (used to calculate expected return/normal return)
# 60 days before the start of the event window 
df.eventsmodel$est_start <- df.eventsmodel$tb3 - estimation.period + 1 # estimation period start date 
df.eventsmodel$est_end <- df.eventsmodel$tb3 - 1 # estimation period end date 

# print events with event windowing 
rownames(df.eventsmodel) <- NULL # remove row numbering 
df.eventsmodel
```

#### Foreign Exchange Market


1. <b><i>GBP/EUR Price</i></b>

<i>Null Hypothesis</i>

- $H_0$: There is no significant difference between the abnormal mean returns for GBP/EUR before a major brexit event and after it. (${\mu_0 = \mu_1}$)

<i>Alternative Hypothesis</i>

- $H_1$: There is a remarkable difference between the abnormal mean returns for GBP/EUR before a major brexit event and after it. (${\mu_0 \neq \mu_1}$) 

```r
####################################################################
# Compute Abnormal Returns for each T in the event window (GBP/EUR)
####################################################################

# first we make copy of the events model dataframe,
# as we will be adding values to specific to this test 
# only get columns needed for this test 
df.eventsgbpeur <- data.frame(df.eventsmodel[c("event", 
                                               "est_start", 
                                               "est_end", 
                                               "event_date", 
                                               "tb3",
                                               "tb2", 
                                               "tb1",
                                               "ta1", 
                                               "ta2",
                                               "ta3")])

# for each event we must compute the expected return/normal return (ER) 
# from the log returns, from and to the estimation time period,
# we will use the mean for the log returns during that period as the ER
df.eventsgbpeur$er <- apply(df.eventsgbpeur[,c("est_start","est_end")], 1, function(x) {
    mean(with(df.logfxrates, df.logfxrates[(as.Date(df.logfxrates$date) >= x["est_start"] 
                                                       & as.Date(df.logfxrates$date) <= x["est_end"]), ])$GBPEUR)
})

# now we need to align the dates for t-3 to t+3 with the trading dates 
# to do so first check if t0/event date was a trading day if yes set to t0 
# if not set t0 date to the first trading date after the event 
# get index of t0 from the log returns dataset 
index.t0 <- apply(df.eventsgbpeur[,c("event", "event_date")], 1, function(x) {
     tmp.t0 <- row.names(head(df.logfxrates[(as.Date(df.logfxrates$date) >= x["event_date"]), ], 1))
})

# now since we have the index for t0 we can easily move to t+1 or t-1 by adding 
# or subtracting to that index 
index.tb3 <- as.numeric(index.t0) - 3 # index at t-3 (to get pre-event log return)
index.tb2 <- as.numeric(index.t0) - 2 # index at t-2 (to get pre-event log return)
index.tb1 <- as.numeric(index.t0) - 1 # index at t-1 (to get pre-event log return)
index.ta1 <- as.numeric(index.t0) + 1 # index at t+1 (to get post-event log return)
index.ta2 <- as.numeric(index.t0) + 2 # index at t+2 (to get post-event log return)
index.ta3 <- as.numeric(index.t0) + 3 # index at t+3 (to get post-event log return)

# next step is to calculate the abnormal returns AR, 
# by subtracting the actual return by the ER for each t in the event window (R - ER)
df.eventsgbpeur$tb3_ar <- df.logfxrates[index.tb3,]$GBPEUR - df.eventsgbpeur$er 
df.eventsgbpeur$tb2_ar <- df.logfxrates[index.tb2,]$GBPEUR - df.eventsgbpeur$er 
df.eventsgbpeur$tb1_ar <- df.logfxrates[index.tb1,]$GBPEUR - df.eventsgbpeur$er 
df.eventsgbpeur$t0_ar  <- df.logfxrates[as.numeric(index.t0),]$GBPEUR - df.eventsgbpeur$er 
df.eventsgbpeur$ta1_ar <- df.logfxrates[index.ta1,]$GBPEUR - df.eventsgbpeur$er 
df.eventsgbpeur$ta2_ar <- df.logfxrates[index.ta2,]$GBPEUR - df.eventsgbpeur$er 
df.eventsgbpeur$ta3_ar <- df.logfxrates[index.ta3,]$GBPEUR - df.eventsgbpeur$er 

##################################################################
# Test hypothesis for GBP/EUR for each event shown previously 
##################################################################

# create data frame for results 
df.resultsgbpeur <- df.eventsmodel[c("event_desc")] 
rownames(df.resultsgbpeur) <-  df.eventsmodel$event

# apply a t-test for each event 
t.summariesgbpeur <- apply(df.eventsgbpeur, 1, function(x) {
    eventar.pre <- as.numeric(c(x["tb3_ar"], x["tb2_ar"], x["tb1_ar"]))
    eventar.post <- as.numeric(c(x["ta1_ar"], x["ta2_ar"], x["ta3_ar"]))
    t.summary <- t.test(eventar.pre, eventar.post, var.eq=TRUE)
    tidy(t.summary)
})

# bind all the results into one dataframe
t.summarygbpeur <- bind_rows(t.summariesgbpeur)

# get results for each event 
df.resultsgbpeur$t_value <- round(t.summarygbpeur$statistic, 4)
df.resultsgbpeur$p_value <- round(t.summarygbpeur$p.value, 4)
df.resultsgbpeur$conf_low <- round(t.summarygbpeur$conf.low, 4)
df.resultsgbpeur$conf_high <- round(t.summarygbpeur$conf.high, 4)
df.resultsgbpeur$pre_aar <- round(t.summarygbpeur$estimate1, 4)
df.resultsgbpeur$post_aar <- round(t.summarygbpeur$estimate2, 4)

# print/show hypothesis testing for GBP/EUR
df.resultsgbpeur
```

- $H_1$: There is a remarkable difference between the abnormal mean returns for GBP/JPY before a major brexit event and after it. (${\mu_0 \neq \mu_1}$) 

```r
####################################################################
# Compute Abnormal Returns for each T in the event window (GBP/JPY)
############################returns########################################

# first we make copy of the events model dataframe,
# as we will be adding values to specific to this test 
# only get columns needed for this test 
df.eventsgbpjpy <- data.frame(df.eventsmodel[c("event", 
                                               "est_start", 
                                               "est_end", 
                                               "event_date", 
                                               "tb3",
                                               "tb2", 
                                               "tb1",
                                               "ta1", 
                                               "ta2",
                                               "ta3")])

# for each event we must compute the expected return/normal return (ER) 
# from the log returns, from and to the estimation time period,
# we will use the mean for the log returns during that period as the ER
df.eventsgbpjpy$er <- apply(df.eventsgbpjpy[,c("est_start","est_end")], 1, function(x) {
    mean(with(df.logfxrates, df.logfxrates[(as.Date(df.logfxrates$date) >= x["est_start"] 
                                                       & as.Date(df.logfxrates$date) <= x["est_end"]), ])$GBPJPY)
})

# next step is to calculate the abnormal returns AR, 
# by subtracting the actual return by the ER for each t in the event window (R - ER)
# now since dates for all fx rates are aligned we can use the previous indices
df.eventsgbpjpy$tb3_ar <- df.logfxrates[index.tb3,]$GBPJPY - df.eventsgbpjpy$er 
df.eventsgbpjpy$tb2_ar <- df.logfxrates[index.tb2,]$GBPJPY - df.eventsgbpjpy$er 
df.eventsgbpjpy$tb1_ar <- df.logfxrates[index.tb1,]$GBPJPY - df.eventsgbpjpy$er 
df.eventsgbpjpy$t0_ar  <- df.logfxrates[as.numeric(index.t0),]$GBPJPY - df.eventsgbpjpy$er 
df.eventsgbpjpy$ta1_ar <- df.logfxrates[index.ta1,]$GBPJPY - df.eventsgbpjpy$er 
df.eventsgbpjpy$ta2_ar <- df.logfxrates[index.ta2,]$GBPJPY - df.eventsgbpjpy$er 
df.eventsgbpjpy$ta3_ar <- df.logfxrates[index.ta3,]$GBPJPY - df.eventsgbpjpy$er 

##################################################################
# Test hypothesis for GBP/JPY for each event shown previously 
##################################################################

# create data frame for results 
df.resultsgbpjpy <- df.eventsmodel[c("event_desc")] 
rownames(df.resultsgbpjpy) <-  df.eventsmodel$event

# apply a t-test for each event 
t.summariesgbpjpy <- apply(df.eventsgbpjpy, 1, function(x) {
    eventar.pre <- as.numeric(c(x["tb3_ar"], x["tb2_ar"], x["tb1_ar"]))
    eventar.post <- as.numeric(c(x["ta1_ar"], x["ta2_ar"], x["ta3_ar"]))
    t.summary <- t.test(eventar.pre, eventar.post, var.eq=TRUE)
    tidy(t.summary)
})

# bind all the results into one dataframe
t.summarygbpjpy <- bind_rows(t.summariesgbpjpy)

# get results for each event 
df.resultsgbpjpy$t_value <- round(t.summarygbpjpy$statistic, 4)
df.resultsgbpjpy$p_value <- round(t.summarygbpjpy$p.value, 4)
df.resultsgbpjpy$conf_low <- round(t.summarygbpjpy$conf.low, 4)
df.resultsgbpjpy$conf_high <- round(t.summarygbpjpy$conf.high, 4)
df.resultsgbpjpy$pre_aar <- round(t.summarygbpjpy$estimate1, 4)
df.resultsgbpjpy$post_aar <- round(t.summarygbpjpy$estimate2, 4)

# print/show hypothesis testing for GBP/JPY
df.resultsgbpjpy
```
